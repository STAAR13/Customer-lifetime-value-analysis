## RFM Analysis

```python
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days = 1)
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue': 'sum'
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
fig, ax = plt.subplots(1,3,figsize=(18,5))
sns.histplot(rfm['Recency'],bins=30, kde=True, ax=ax[0]).set(title='Recency Distribution')
sns.histplot(rfm['Frequency'],bins=30, kde=True, ax= ax[1]).set(title='Frequency Distribution')
sns.histplot(rfm['Monetary'], bins=30, kde=True, ax = ax[2]).set(title='Monetary Distribution')
plt.show()
```

#### Recency, Frequency, Monetary (RFM) metrics computed:

**Recency**: Days since last purchase.

**Frequency**: Total purchase transactions.

**Monetary**: Total spend per customer.

```python
rfm['R_quartile'] = pd.qcut(rfm['Recency'],4,labels=[4,3,2,1])
rfm['F_quartile'] = pd.qcut(rfm['Frequency'].rank(method='first'),4,labels=[1,2,3,4])
rfm['M_quartile'] = pd.qcut(rfm['Monetary'],4,labels=[1,2,3,4])
rfm['RFM_score'] = rfm['R_quartile'].astype(int) + rfm['F_quartile'].astype(int) + rfm['M_quartile'].astype(int)
```

```python
high_value = rfm[rfm['RFM_score']>=9]
print(f"High-value customers count: {high_value.shape[0]}")
```

#### Customers were scored into quartiles:

Higher scores indicate more recent, frequent, and higher-value customers.

```python
# Step 8: Customer Segmentation based on CLTV percentiles
summary['cltv_segment'] = pd.qcut(summary['cltv'], q=[0, 0.2, 0.8, 1.0], labels=['Low', 'Medium', 'High'])

# View distribution
print(summary['cltv_segment'].value_counts())

# View example customers from each segment
summary.groupby('cltv_segment')[['cltv', 'predicted_purchases', 'expected_avg_profit']].mean()
```

# Key Findings and Recommendations

## 1. Revenue by Country

- **United Kingdom** generates the highest share of revenue (**over 80%** of total), followed by:
  - Netherlands  
  - Ireland  
  - Germany  
  - France  

**Recommendation:**  
Continue to prioritize marketing and inventory for the **UK**, but also explore growth opportunities in underrepresented yet promising markets such as **Germany** and **France**.

---

## 2. Monthly Revenue Trends

- **Peak in November**: Significant revenue spike, likely due to **Black Friday** and **holiday season** sales.  
- **Dip in December**: Notable drop, potentially due to **inventory constraints** or **seasonal demand** shifts.

**Recommendation:**  
Optimize **Q4 promotions** and **inventory planning** to:
- Fully leverage peak demand periods.
- Minimize lost revenue opportunities.

---

## 3. Customer Segmentation

- **Top 20%** of customers contribute **~80% of the revenue** (Pareto Principle).  
- **Bottom 20%** show low purchase frequency and spending.

**Recommendations:**
- **High-Value Customers**: Retain and nurture through:
  - Loyalty programs  
  - Early access to sales  
  - Exclusive perks  

- **Low-Value Customers**: Re-engage using:
  - Targeted discounts  
  - Curated product recommendations  

---

## 4. Customer Lifetime Value (CLTV) Analysis

### Disproportionate Revenue Contribution:
CLTV modeling reveals:
- A small group of customers generates a major share of **predicted future revenue**.
- Even customers with **low purchase counts** can have **high value** due to large order sizes.

### Segmentation Insights:

#### High CLTV Segment:
- **Highest predicted future value**
- Large order amounts and/or steady engagement  
- Most critical group for **personalized retention** and **upselling**

#### Medium CLTV Segment:
- Moderate but promising value  
- Target for **incremental growth** via **cross-selling** and **engagement**

#### Low CLTV Segment:
- Low predicted spend and frequency  
- **Marketing efforts** should be **carefully measured** to ensure ROI
